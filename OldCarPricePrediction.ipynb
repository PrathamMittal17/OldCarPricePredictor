Dataset taken from kaggle : 
https://www.kaggle.com/datasets/avikasliwal/used-cars-price-prediction 

Name : The brand and model of the car.

Location : The location in which the car is being sold or is available for purchase.

Year: The year or edition of the model.

Kilometers_Driven: The total kilometres driven in the car by the previous owner(s) in KM.

Fuel_Type: The type of fuel used by the car. (Petrol, Diesel, Electric, CNG, LPG)

Transmission: The type of transmission used by the car. (Automatic / Manual)

Owner_Type: First Hand/Second Hand

Mileage: The standard mileage offered by the car company in kmpl or km/kg

Engine: The displacement volume of the engine in CC.

Power: The maximum power of the engine in bhp.

Seats: The number of seats in the car.

New_Price: The price of a new car of the same model.

Price : Price at which car got sold.














```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
train_data = pd.read_csv("train-data.csv",index_col='Unnamed: 0')
train_data.head()
```

```python
train_data.shape
```

```python
train_data.isnull().sum()
```

### Value counts of owner type

```python
train_data['Owner_Type'] = np.where(train_data['Owner_Type'] == 'Fourth & Above','Third & Above',train_data['Owner_Type'])
train_data['Owner_Type'] = np.where(train_data['Owner_Type'] == 'Third','Third & Above',train_data['Owner_Type'])
train_data['Owner_Type'].value_counts()
```

### Dealing with fuel type

```python
train_data.drop(train_data[(train_data['Fuel_Type']=='LPG') | (train_data['Fuel_Type']=='Electric')].index ,inplace=True)
train_data.Fuel_Type.value_counts()
```

### Dealing with Seats

```python
train_data.drop(train_data[(train_data['Seats']==0) | (train_data['Seats']>8)].index ,inplace=True)
```

### Dealing with Engine

```python
train_data['Engine'] = train_data['Engine'].str.split(expand=True)[0]
```

### Dealing with Power

```python
train_data['Power'] = train_data['Power'].str.split(expand=True)[0]
train_data.Power.replace('null',np.nan,inplace=True)
```

### Dealing with Mileage

```python
train_data['Mileage'] = train_data['Mileage'].str.split(" ",expand=True)[0]
```

### Dropping Null

```python
train_data.drop('New_Price',inplace=True,axis=1)
train_data.dropna(inplace=True)
train_data.reset_index(drop=True,inplace=True)
```

```python
print(train_data.isnull().sum(),"\nShape: " ,train_data.shape)
```

### Dealing with Name column

```python
train_data['Brand'] = train_data['Name'].str.lower().str.split(" ",expand=True)[0].str.title()
```

```python
train_data.Brand.unique()
```

```python
train_data.Brand = np.where(train_data.Brand=='Land','Land Rover',train_data.Brand)
```

```python
x =  train_data['Name'].str.lower().str.split(" ")
models = []
for i in x:
    if len(i[1:3][0]) == 1:
        model = ' '
        model += i[1:3][0] + " " + i[1:3][1]
        models.append(model.title().strip())
    else:
        models.append(i[1].title())

train_data['Model'] = models
train_data['Model'] = np.where(train_data['Model']=='Wagon','WagonR',train_data['Model'])
train_data.drop('Name',inplace=True,axis=1)
```

### Changing Datatypes

```python
train_data.dtypes
```

```python
train_data['Mileage'] = train_data['Mileage'].astype(float)
train_data['Engine'] = train_data['Engine'].astype(int)
train_data['Power'] = train_data['Power'].astype(float)
train_data['Seats'] = train_data['Seats'].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
X = train_data.drop(['Price'],axis=1)
y = train_data['Price']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
le = LabelEncoder()
unique_model = X['Model'].unique()
le.fit_transform(unique_model)
```

```python
cat_features = ['Fuel_Type','Transmission','Owner_Type','Location','Brand']
ohe = make_column_transformer((OneHotEncoder(),cat_features),remainder="passthrough")
```

```python
X_train['Model'] = le.transform(X_train['Model'])
X_test['Model'] = le.transform(X_test['Model'])
```

```python
rfr = Pipeline([("ohe",ohe),("model",RandomForestRegressor())])
knn = Pipeline([("ohe",ohe),("ss",StandardScaler(with_mean=False)),("model",KNeighborsRegressor())])
lr = Pipeline([("ohe",ohe),("ss",StandardScaler(with_mean=False)),("model",LinearRegression())])

models = [rfr,knn,lr]

for model in models:
    model.fit(X_train,y_train)
    y_preds = model.predict(X_test)
    print(model['model'],  "\n R2 SCORE: " ,r2_score(y_test,y_preds),"\n MSE: " ,mean_squared_error(y_test,y_preds))
```

```python
import pickle
pickle.dump(rfr, open("model.pkl", 'wb'))
pickle.dump(le, open("le.pkl", 'wb'))
```
